#Characterizing Failure in Manufacturing Process
##The Case for Repairability
_Daniel Cellucci and Kenneth C. Cheung_

###Motivation

Our motivation behind this brief theoretical study is to examine whether repairability is an effective strategy for manufacturing in space.

Systems such as SpiderFab's Trussembler and Intelligent Precision Jigging Robots use non-reversible processes of assembly. The justification for this approach is that reversible processes add such complexity to the overall system that the inclusion of this feature is not justified.

We will examine the truth of this hypothesis by characterizing the act of building a structure as a **manufacturing process** that is either non-repairable, quasi-repairable, or completely repairable.

###A manufacturing process

A **manufacturing process** consists of a sequence of $n$ identical steps, each of which has a probability of **catastrophic failure** $p$.

We define **catastrophic failure** as reducing the performance of the overall system to such an extent that it must be repaired or thrown out.

There are three different manufacturing strategies according to this scheme. 
 
 1. **Non-repairable** systems are configured in such a way that a failure results in the total scrapping of the product. This involves systems where the end product is a monolithic unit.
 2. **Quasi-repairable** systems can allow a certain finite number of failures before the system can be scrapped. This includes systems with rudimentary repair, or with those where the end product is a series of modular units combined in a irreversible way.
 3. **Repairable** systems are configured in such a way that the failure of each step is localized and reversible, resulting in the waste of just the step that failed.

Since we assign one of two states to each step in a process, The resulting probability distribution follows that of the Binomial Distribution.

$$
f(k;n;p) = {n \choose k}p^k(1-p)^{n-k}
$$

Where, in the case of a **non-repairable** system, $k = 0$, and in the case of **quasi-repairable**, $k > 0$, where the specific value is dependent on the system in question.

Fully **repairable** systems don't really follow these rules. The details are included below.

###The Non-Repairable System

In the case of $k=0$, the probability of successfully creating a product with a process consisting of $n$ steps and a per-step probability of catastrophic failure $p$ simplifies to

$$
f(0;n,p) = {n\choose0} p^0 (1-p)^{n-0} = (1-p)^n
$$

```python
#the array of p values we want to examine
numprob = 3
p = [0.001,0.005,0.01]
#corresponding colors of the graphs that are produced
colo = ['r','b','g']

#All of our data of interest
max_num = 101

prob_of_product = []
for i in range(0,numprob):
    prob_of_product.append([])
    for n in range(0,max_num):
        prob_of_product[i].append((1-p[i])**n)
    c = colo[i]
    plot(range(0,max_num),prob_of_product[i],c,label="p ="+str(p[i]))
ylabel('Probability of a perfect (k = 0) product')
xlabel('Number of manufacturing steps ($n$)')
legend(loc=3)
show()
```

This shows the relative decrease in probability of success as the size of the system is increased for three different failure probabilities. 

However, this does not fully encapsulate the **waste** generated by a manufacturing process. By **waste** we mean the overall number of steps (and therefore resources) expended that did not end up going into the final product. 

There are two strategies for reacting to a catastrophic failure that define the upper and lower bounds for the amount of waste generated.

###Strategies for Reacting to Errors

In the event of catastrophic failure of a system, there are two possible strategies that represent the **best** and **worst** case scenarios, and are related to the ability of the system to assess the functionality of the product.

 1. The **worst case scenario** occurs when the system cannot assess the functionality of the product until after the entire manufacturing process has been performed. It therefore must repeat the entire process until it produces a product that has below the threshold number of errors. 
 
 2. The **best case scenario** occurs when the system can ascertain the success of a step immediately after performing it, and can throw out the result when the number of failures reaches above a certain amount. 

**Worst Case Waste**

The probability distribution for the number of possible repetitions of a full manufacturing process before success is encapsulated in a _negative binomial distribution_. 

In the case of a **non-repairable** system, the probability of a successful product is $q = (1-p)^n$, and therefore the expected number of failures before $t$ successes is the mean of the negative binomial distribution

$$
\frac{t(1-q)}{q}
$$

Each failure produces a waste of $n$ steps, and $t=1$ is the target number of successes. Therefore,

$$
W_{wc} = n\frac{1-(1-p)^n}{(1-p)^n}
$$

```python
#the array of p values we want to examine
numprob = 3
p = [0.001,0.005,0.01]
#corresponding colors of the graphs that are produced
colo = ['r','b','g']

#All of our data of interest
max_num = 101

waste_wc = []
for i in range(0,numprob):
    waste_wc.append([])
    for n in range(0,max_num):
        waste_wc[i].append(n*(1-(1-p[i])**n)/(1-p[i])**n)
    c = colo[i]
    plot(range(0,max_num),waste_wc[i],c,label="p ="+str(p[i]))
ylabel('Waste Generated in the Worst Case')
xlabel('Number of manufacturing steps ($n$)')
legend(loc=2)
show()
```

As one can see, there is an exponential increase in the amount of waste generated ($W_{wc}$) as the number of manufacturing steps increases. 

**Best Case Waste**

In the best case, we are instead looking at a slightly more complex set of behaviors, so I'll detail them below. 

 1. A process starts
 2. A step is performed, with probability $p$ of failure
 3. `If` the number of steps $m > n$ (the maximum number) GOTO 6.
 4. `else-if` p is successful, `GOTO` 2.
 5. `else` (p is unsuccessful), `GOTO` 1.
 6. Finished product.
 
As a result, when the process reaches 6., the distribution of waste will also follow a negative binomial distribution, albeit with different parameters. 

If $p$ is the probability of failure for a manufacturing process of $n$ steps with $r=1$ allowed errors, then the distribution of waste for a discard of size $k$ is

$$
f(k;r=1,p) = \binom{k}{k}p^1(1-p)^k = p\ (1-p)^k
$$

However, there is an additional requirement, $k \leq n$, since the process stops when the target size is reached. This results in a truncation of the distribution:

$$
f(k|K\leq n) = \frac{g(k)}{F(n)}
$$

where $g(k) = f(k)$ for all $k \leq n$ and $g(k) = 0$ otherwise, and $F(k)$ is the cumulative density function for $f(k)$. Thus, 

$$
f(k|K\leq n; r,p) = \frac{1}{I_p(1,n+1)}p\ (1-p)^k
$$

A factor for the waste produced by the process is the mean of this distribution, which, because of the truncation, does not have a neat analytical form. 

$$
\bar{W_{bc}} = \frac{1}{I_p(1,n+1)}\sum_{k=0}^nk\ p\ (1-p)^k
$$

To get the total waste, this factor must be multiplied by the ratio of the failure probability to the success probability

$$
W_{bc} = \frac{1-(1-p)^n}{(1-p)^n}\frac{1}{I_p(1,n+1)}\sum_{k=0}^nk\ p\ (1-p)^k 
$$

```python
numprob = 3
p = [0.001,0.005,0.01]
#corresponding colors of the graphs that are produced
colo = ['r','b','g']

#All of our data of interest
max_num = 101

waste_bc = []
for i in range(0,numprob):
    waste_bc.append([])
    for n in range(0,max_num):
        sum = 0
        for k in range(0,n+1):
            sum = sum + k*p[i]*(1-p[i])**k
        waste_bc[i].append(sum/sp.betainc(1,n+1,p[i])*(1-(1-p[i])**n)/(1-p[i])**n)
    c = colo[i]
    plot(range(0,max_num),waste_bc[i],c,label="p ="+str(p[i]))
ylabel('Waste Generated in the Best Case')
xlabel('Number of manufacturing steps ($n$)')
legend(loc=2)
show()

for i in range(0,numprob):
    c = colo[i]
    plot(range(0,max_num), waste_bc[i],c,label="p ="+str(p[i]))
    plot(range(0,max_num), waste_wc[i],c+'--')

ylabel('Waste Generated')
xlabel('Number of manufacturing steps ($n$)')
legend(loc=2)
```

Above is the full envelope of possible wasted resources, given the worst case and best case scenarios. 

###The Quasi-Repairable System

If we have a quasi-repairable system, we can allow up to $r$ errors to occur before failure. 

Therefore, for the defined process above, the probability of a successful product that has $k \leq r$ errors can be ascertained from a bimodal distribution. 

$$
f(k;n;p) = {n \choose k}p^k(1-p)^{n-k}
$$

For non-zero values of $r$, we seek to produce anything that has less than $r$ errors. Therefore, the probability of creating a product with any number of errors $k \leq r$ is the cumulative binomial distribution:

$$
I_{(1-p)}(n-r, 1+r) = \sum^{n}_{j=n-r}\binom{n}{j}(1-p)^j p^{n-j}
$$

The question is, then, given a process of $n$ steps and a number of allowed errors $r$ (where $r \leq n$), what is the resulting probability surface $P(n,r)$?



```python
num = [50,100,200]
p = 0.01

prob_surf = []

for i in range(0,len(num)):
    prob_surf.append([])
    n = num[i]
    for r in range(0,n):
        prob_surf[i].append(sp.betainc(n-r,1+r,1-p))
    plot(range(0,n),prob_surf[i], label="$n = $" + str(num[i]))

ylim([0,1])
xlim([0,10])
xlabel("Number of allowed errors $r$")
ylabel("Probability $P$")
legend()
title("# Allowed errors $r$ vs. prob of process success $P$\n for per-step prob. $p$ = " + str(p) + " of failure.")
```

A few notes on this chart. 

The value corresponding to a $P = 0.5$ probability of success is the mean of the binomial distribution, $r = np$.

The variance is also straightforward and useful $\sigma^2 = np(1-p)$, since it gives a sense of how the responsiveness to $r$ scales with $n$ and $p$. For instance, the value of $r$ which produces a probability of success above a certain threshold $P_{thr}$ is

$$
    r = np + \sqrt{np(1-p)}\sqrt{\frac{1}{2P_{thr}-1}} = np+\sqrt{\frac{np(1-p)}{2P_{thr}-1}}
$$

so, for $n = 200$, $p = 0.1$, and $P_{thr} = 0.9$, $r > 24$.

**Worst Case Waste**

Here, we take the previously calculated probability of success $q$ for a process of $n$ steps with failure probability $p$ and number of allowable errors $r$.

$$
q = I_{(1-p)}(n-r, 1+r)
$$

We then want to look at how many failures were necessary to produce $t$ successes. The mean of the negative binomial distribution gives us just that,

$$
\frac{t(1-q)}{q}
$$

In this case, our process length is $n$ and $t=1$ is our target number of successes, so the total number of steps wasted to produce one result is:

$$
W_{wc} = n\frac{(1-q)}{q}
$$

substituting for $q = I_{(1-p)}(n-r, 1+r)$:

$$
W_{wc} = n\frac{(1-I_{1-p}(n-r,1+r)}{I_{1-p}(n-r,1+r)} = n\frac{I_p(1+r,n-r)}{I_{1-p}(n-r,1+r)}
$$

```python
#Calculate the worst-case waste for non-repairable or quasi-non repairable systems

n = 100
p = 0.01

waste_wc = []

for r in range(0,n):
    val = n*sp.betainc(1+r,n-r,p)/sp.betainc(n-r,1+r,1-p)
    waste_wc.append(val)
    
plot(range(0,n),waste_wc)

lb = 0
xlim([lb,4])
ylim([0,waste_wc[lb]])
#print(waste_wc[20])
```

###Aside: The Regularized/Incomplete Beta Function

$$
B(a,b)= \int_0^1 t^{a-1}(1-t)^{y-1}\textrm{d}t
$$

$$
B(x;a,b) = \int_0^x t^{a-1}(1-t)^{b-1}\textrm{d}t
$$

$$
I_x(a,b) = \frac{B(x;a,b)}{B(a,b)}=\sum_{j=a}^{a+b-1}\binom{a+b-1}{j}x^j(1-x)^{a+b-1-j}
$$
