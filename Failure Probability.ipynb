##A manufacturing process

A process consists of a sequence of identical steps, each of which has a probability of catastrophic failure $p$.

We define catastrophic failure as reducing the performance of the overall system to such an extent that it must be repaired or thrown out.

Therefore, the probability of a functioning product at the end of a process that is $n$ steps long is a binomial distribution where $k=0$. 

```python
#the array of p values we want to examine
numprob = 3
p = [0.001,0.005,0.01]
#corresponding colors of the graphs that are produced
colo = ['r','b','g']

#All of our data of interest
max_num = 101
```

```python
prob_of_product = []
for i in range(0,numprob):
    prob_of_product.append([])
    for n in range(0,max_num):
        prob_of_product[i].append((1-p[i])**n)
    c = colo[i]
    plot(range(0,max_num),prob_of_product[i],c+'o')
ylabel('Probability of a perfect (k = 0) product')
xlabel('Number of required manufacturing steps ($n$)')
show()
```

There are two bounds on testing of the product to ascertain if an error has occured:

  1. In the upper bound, we test the product at the end of the process, and build a new one if the current one has a catastrophic error
  2. In the lower bound, we test the success of each step when it is performed, and restart if there is a catastrophic error
  
The question we seek to answer is: **How much waste is generated by each of these approaches?**

###The upper bound
In the upper bound, we repeat an $n$-step manufacturing process until we reach a success. The question is, how many failures did it take to get there? 

This is encapsulated by a _negative binomial distribution_, where $r=1$ and $p$ is the probability of producing a perfect result from an $n$-step process. The mean $\bar{f}$ number of failures of a negative binomial distribution with probability $p$ that stops after $r$ successes is 

$$
\bar{f}=\frac{pr}{1-p}
$$

thus, if we take the number of manufacturing steps $n$ as a measure of the resources (time and material) expended to produce an object, the amount of waste $W$ produced in the course of creating one product using this process is

$$
W = n\bar{f}_{r=1} = n\frac{p}{1-p}
$$


```python
#Upper bound waste arrays - percent and absolute
ub_waste_per = []
ub_waste_abs = []


for i in range(numprob):
    ub_waste_per.append([])
    ub_waste_abs.append([])
    for n in range(0,max_num):
        ub_waste_per[i].append((1-prob_of_product[i][n])/prob_of_product[i][n])
        ub_waste_abs[i].append((1-prob_of_product[i][n])/prob_of_product[i][n]*n)
    c = colo[i]
    plot(range(0,max_num), ub_waste_per[i],c+'o')
xlabel('Number of steps ($n$)')
ylabel('Average number of $n$-step processes wasted')
show()

for i in range(numprob):
    c = colo[i]
    plot(range(0,max_num), ub_waste_abs[i],c+'o')
xlabel('Number of steps ($n$)')
ylabel('Average number of steps wasted ($n$)')
print(ub_waste_abs[2][100])
```

###The Lower Bound

The other, and best case option, is to check after every step in the $n$-step process, and start over if there is a catastrophic error. 

This can also be represented by a negative binomial distribution. If we decide to repeat a manufacturing step until we produce one catastrophic error, then the distribution of lengths of consecutive successful steps is simply a negative binomial distribution with $r=1$ and where $p$ is the probability of failure for the process.  

```python
r = 1

pdf_neg_binom = []
cdf_neg_binom = []


for i in range(0,numprob):
    pdf_neg_binom.append([])
    cdf_neg_binom.append([])
    tsum = 0
    for n in range(0,max_num):
        tval = ((p[i])**r)*((1-p[i])**n)
        #tval = ((1-p[i])**n)
        pdf_neg_binom[i].append(tval)
        tsum = tsum + tval
        cdf_neg_binom[i].append(tsum)
    c = colo[i]
    plot(range(0,max_num),pdf_neg_binom[i],c+'o')

title("PDF for Negative Binomial Distribution with r = 1")
show()

for i in range(0,numprob):
    c = colo[i]
    plot(range(0,max_num),cdf_neg_binom[i],c+'o')

title("CDF for Negative Binomial Distribution with r = 1")
show()
```

This means that, after one catastrophic failure, the products that are the result of $n'$ manufacturing steps fall on the above probability distribution. 

However, there is an additional condition on the lower bound process- the manufacturing process doesn't go until a failure but stops after $n$ steps.

As a result, if a given process is repeated until it produces $n$ successes in a row, the distribution of previous strings of consecutive successes is the distribution of waste for the process. This distribution is found by taking the _truncated negative binomial distribution_.

The truncated distribution for a process that terminates after $n$ steps is

$$
t(x) = f(x|X\leq n) = \frac{g(x)}{F(n)}
$$

Where $g(x) = f(x)$ if $x\leq n$ and $g(x)=0$ everywhere else, and $F(x)$ is the cumulative distribution

```python
#of n percent chance of failure, the distribution is PDF for neg binomial

trunc_num = 100

tdf_neg_binom = []

for i in range(0,numprob):
    tot = cdf_neg_binom[i][trunc_num-1]
    tdf_neg_binom.append([])
    for n in range(0,trunc_num):
        tdf_neg_binom[i].append(pdf_neg_binom[i][n]/tot)
    c = colo[i]
    plot(range(0,trunc_num),tdf_neg_binom[i],c+'o')
title('Normalized Distribution of waste for $N='+str(trunc_num)+'$')
xlabel('Number of manufacturing steps wasted ($n$)')
ylabel('Probability')
show()


histogram = []
notfail = True
for i in range(500):
    histogram.append(0)
'''
for i in range(0,1000000):
    size = 0
    notfail = True
    while notfail and size < 500:
        if(rand()<0.01):
            notfail = False
            histogram[size] = histogram[size]+1/1000000
        else:
            size = size + 1
newhist = []
for i in range(500):
    newhist.append(histogram[i]/tdf_neg_binom[2][i])
            
plot(range(0,500),newhist)
#plot(range(0,trunc_num),tdf_neg_binom[1])
'''
```

The total waste for a given process consisting of $N$ steps is therefore the sum of the number of manufacturing steps wasted, weighted by the probability of a failure occuring after that many steps. That is, 

$$
T = \sum_{n=0}^{N-1} (n+1)\ t(n)
$$

The reason for $N-1$ and $n+1$ is that we can only check if a process is successful after the fact. 

We then want to take this total waste and multiply it by the probability of failing to produce a perfect product. This is from the first chart we generated.

```python
lb_waste = []
lb_waste_proper = []
lb_waste_norm = []

for i in range(0,numprob):
    lb_waste.append([])
    lb_waste_proper.append([])
    lb_waste_norm.append([])
    for n in range(1,max_num):
        tot = cdf_neg_binom[i][n-1]
        t_sum = 0
        for s in range(0,n):
            t_sum = t_sum + pdf_neg_binom[i][s]/tot*(s+1)
        lb_waste[i].append(t_sum)
        lb_waste_proper[i].append(t_sum*(1-prob_of_product[i][n])/prob_of_product[i][n])
        lb_waste_norm[i].append(t_sum*(1-prob_of_product[i][n])/(n*prob_of_product[i][n]))

for i in range(0,numprob):    
    c = colo[i]
    plot(range(0,max_num-1),lb_waste[i],c+'o')
show()

for i in range(0,numprob):    
    c = colo[i]
    plot(range(0,max_num-1),lb_waste_proper[i],c+'o')
show()

for i in range(0,numprob):    
    c = colo[i]
    plot(range(0,max_num-1),lb_waste_norm[i],c+'o',label=str(p[i]))

title('Waste factor as a function of beam length')
xlabel('Beam length $L$')
ylabel('# of length $L$ beams wasted')
legend(loc=4)
show()
```

```python
window = 7
#waste factor bounds
for i in range(0,numprob):
    temp = []
    temp_ub = []
    temp_lb = []
    for w in range(window-1):
        temp.append(w*p[i])
        temp_ub.append(ub_waste_abs[i][w])
        temp_lb.append(lb_waste_proper[i][w])
    c = colo[i]
    plot(range(window-1),temp_ub,c+'o')
    plot(range(window-1),temp_lb,c+'^')
    plot(range(window-1),temp,c)
```
